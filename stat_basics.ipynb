1: операция?

2: Возраст

3: ректальная температура

- Повышенная температура может возникнуть из-за инфекции.
- температура может быть снижена, когда животное находится в состоянии позднего шока
- нормальная температура составляет 37,8

4: пульс

- линейный

5: частота дыхания

7: боль - субъективное суждение об уровне боли лошади

- В общем, чем больнее, тем больше вероятность того, что это потребует хирургия
- предварительное лечение боли может в некоторой степени маскировать уровень боли

8: результат-что в конце концов случилось с лошадью?

```python
q1_rectal_temperature = data2['rectal_temperature'].quantile(0.25)
q3_rectal_temperature = data2['rectal_temperature'].quantile(0.75)
iqr_rectal_temperature = q3_rectal_temperature - q1_rectal_temperature
lb_rectal_temperature = q1_rectal_temperature - (1.5 * iqr_rectal_temperature)
ub_rectal_temperature = q3_rectal_temperature + (1.5 * iqr_rectal_temperature)
remove_outliers1 = data2[data2['rectal_temperature'].between(lb_rectal_temperature, ub_rectal_temperature, inclusive=True)].sort_values('rectal_temperature')
remove_outliers1
```

```python
pd.concat([data2,remove_outliers1]).drop_duplicates(keep=False)
```

```python
q1_pulse = data2['pulse'].quantile(0.25)
q3_pulse = data2['pulse'].quantile(0.75)
iqr_pulse = q3_pulse - q1_pulse
lb_pulse = q1_pulse - (1.5 * iqr_pulse)
ub_pulse = q3_pulse + (1.5 * iqr_pulse)
remove_outliers2 = data2[data2['pulse'].between(lb_pulse, ub_pulse, inclusive=True)].sort_values('pulse')
remove_outliers2
```

```python
pd.concat([data2,remove_outliers2]).drop_duplicates(keep=False)
```

```python
q1_respiratory_rate = data2['respiratory_rate'].quantile(0.25)
q3_respiratory_rate = data2['respiratory_rate'].quantile(0.75)
iqr_respiratory_rate = q3_respiratory_rate - q1_respiratory_rate
lb_respiratory_rate = q1_respiratory_rate - (1.5 * iqr_respiratory_rate)
ub_respiratory_rate = q3_respiratory_rate + (1.5 * iqr_respiratory_rate)
remove_outliers3 = data2[data2['respiratory_rate'].between(lb_respiratory_rate, ub_respiratory_rate, inclusive=True)].sort_values('respiratory_rate')
remove_outliers3
```

```python
pd.concat([data2,remove_outliers3]).drop_duplicates(keep=False)
```

```python
data = pd.read_csv('horse_data.csv', na_values='?')
data.columns=["surgery?", "Age", "Hospital_Number", "rectal_temperature", "pulse", "respiratory_rate", "temperature_of_extremities", "peripheral_pulse", "mucous_membranes", "capillary_refill_time", "pain", "peristalsis", "abdominal_distension", "nasogastric_tube", "nasogastric_reflux", "nasogastric_reflux_PH", "rectal_examination_feces", "abdomen", "packed_cell_volume", "total_protein", "abdominocentesis_appearance", "abdomcentesis_total_protein", "outcome", "surgical_lesion?", "type_of_lesion_1", "type_of_lesion_2", "type_of_lesion_3", "cp_data"]
data_ins = data[["surgery?", "Age", "rectal_temperature", "pulse", "respiratory_rate", "temperature_of_extremities", "pain", "outcome"]]
data2.dropna(thresh=4)
#Оставляем только строки в которых как минимум 4 из 8 строк заполнены.
```

```python
data_ins['surgery?'].fillna(data2['surgery?'].mode()[0], inplace=True)
# так как показатель возраста, маленький процент пропусков, подставляем модальное значение.
```

```python
data_ins['outcome'].fillna(data2['outcome'].mode()[0], inplace=True)
# хочется верить, что животное живо, ну или было живо на момент сбора информации, подставляем модальное значение.
```

```python
# показатель не играет большой роли, заменяем пустые значения медианным значением.
data_ins['respiratory_rate'].fillna(data2['respiratory_rate'].median(), inplace=True)
```

```python
# чем больше частота дыхания, тем выше пульс. Я решил поставить медианное значение в зависимости от частоты дыхания
data_ins['pulse'].fillna(data_ins.groupby('respiratory_rate')['pulse'].transform('median'), inplace=True)
```

```python
# непопавшую в значения медианы строку, заполняем модальным значением.
data_ins['pulse'].fillna(data_ins['pulse'].mode()[0], inplace=True)
```

```python
# боль на начальном этапе, вызывает учащенное сердцебиение, поэтому показатель боли связан с пульсом
data_ins['pain'].fillna(data_ins.groupby('pulse')['pain'].transform('median'), inplace=True)
```

```python
#обычно меняется по мере развития проблемы например, может начаться нормально, 
#затем стать повышенным из-за поражения, возвращаясь в нормальный диапазон, когда лошадь впадает в шок
data_ins['rectal_temperature'].fillna(data_ins.groupby('pain')['rectal_temperature'].transform('median'), inplace=True)
```

```python
data_ins['temperature_of_extremities'].fillna(data_ins.groupby('rectal_temperature')['temperature_of_extremities'].transform('median'), inplace=True)
data_ins['temperature_of_extremities'].fillna(data_ins['temperature_of_extremities'].mode()[0], inplace=True)
data_ins.info()
```
